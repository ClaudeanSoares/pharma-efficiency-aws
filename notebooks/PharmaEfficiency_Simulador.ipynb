```python
# --- PHARMAEFFICIENCY: VALIDATAÇÃO DE INFRAESTRUTURA ---
# Comparativo: Servidor Fixo vs. Arquitetura Serverless AWS

def analisar_viabilidade(vendas_mensais):
    custo_servidor_on_premises = 250.00  # Manutenção, energia e hardware

    # AWS Lambda: Custo por milhão de solicitações é baixíssimo
    # EventBridge: Gerenciamento de eventos
    custo_processamento_aws = (vendas_mensais * 0.0000166) + 1.00

    economia = custo_servidor_on_premises - custo_processamento_aws
    return custo_processamento_aws, economia

# Simulação com 80.000 registros de vendas/mês
vendas = 80000
custo_final, poupanca = analisar_viabilidade(vendas)

print(f"Relatório de Eficiência para Abstergo Industries")
print(f"Custo Cloud: ${custo_final:.2f} | Economia Mensal: ${poupanca:.2f}")
```
